```python
def rotateRandomImage(listOfImg, rotate,prefix,directory,sample):
    filenames = random.sample(listOfImg,sample)
    for img in filenames:
        Original_Image = Image.open(directory+img)
        rotated_image = Original_Image.rotate(rotate)
        split=img.split('.')
        print(directory+split[0]+prefix+'.jpg')
        rotated_image.save(directory+split[0]+prefix+'.jpg')
```

```python
def cropRandomImage(listOfImg,prefix,directory,sample):
    filenames = random.sample(listOfImg,sample)
    for img in filenames:
        Original_Image = Image.open(directory+img)
        cropped = Original_Image.crop((50, 50, 200, 200))
        split=img.split('.')
        print(directory+split[0]+prefix+'.jpg')
        cropped.save(directory+split[0]+prefix+'.jpg')
```

```python
def gaussianBlurRandomImage(listOfImg,prefix,directory,sample):
    filenames = random.sample(listOfImg,sample)
    for img in filenames:
        Original_Image = Image.open(directory+img)
        blurr = Original_Image.filter(ImageFilter.GaussianBlur(radius = 3))
        split=img.split('.')
        print(directory+split[0]+prefix+'.jpg')
        blurr.save(directory+split[0]+prefix+'.jpg')
```

```python
def brightnessRandomImage(listOfImg,prefix,directory,sample):
    filenames = random.sample(listOfImg,sample)
    for img in filenames:
        if 'crop' not in img and 'blur' not in img and 'rot' not in img and 'bri' not in img :
            Original_Image = Image.open(directory+img)
            enhancer = ImageEnhance.Brightness(Original_Image)
            factor = 1.5 #brightens the image
            im_output = enhancer.enhance(factor)
            split=img.split('.')
            print(directory+split[0]+prefix+'.jpg')
            im_output.save(directory+split[0]+prefix+'.jpg')
```
